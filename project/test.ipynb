```python
import requests

datestr = '20230616'
respone = requests.get('https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + datestr + '&type=ALLBUT0999')
```

```python
content = respone.text.replace('=', '')
content
```

```python
lines = content.split('\n')

# print(lines)
for i in lines:
    # 指數類
    if len(i.split('",'))==7:
        split_data = i.split('",')
        # print(split_data)
        data ={}
        data['指數'] = split_data[0].replace('"','')
        data['收盤指數'] = split_data[1].replace('"','')
        data['漲跌'] = split_data[2].replace('"','')
        data['漲跌點數'] = split_data[3].replace('"','')
        data['漲跌百分比'] = split_data[4].replace('"','')
        data['特殊處理註記'] = split_data[5].replace('"','')
        print(data)
# filter(lambda l:len(l.split('",')) > 10, lines)
```
